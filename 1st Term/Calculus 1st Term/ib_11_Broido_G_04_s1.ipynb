```python
# УПРАЖНЕНИЕ 1
import numpy as np
import matplotlib.pyplot as plt

n = 256 + 0j
pi = np.pi
r = np.abs(n)
phi = np.angle(n)
k = np.arange(1,9)
n_roots = r**(1/8) * ( np.cos( (phi + 2.0*pi*k) / 8) +
                  1j * np.sin( (phi + 2.0*pi*k) / 8))
for n,i in enumerate(n_roots):
    n_roots[n] = "{:g}".format(i)
print(n_roots)

left_roots = np.array([n_roots[2], n_roots[3], n_roots[4]])
right_roots= np.array([n_roots[6], n_roots[7], n_roots[0]])

plt.scatter(np.real(n_roots), np.imag(n_roots), marker="+", 
            color="red", s=50)
plt.scatter(np.real(n_roots[3]), np.imag(n_roots[3]), marker="*", 
            color="green", s=200)
plt.scatter(np.real(n_roots[7]), np.imag(n_roots[7]), marker="*", 
            color="green", s=200)

plt.plot(np.real(left_roots), np.imag(left_roots), 'o-.')
plt.plot(np.real(right_roots), np.imag(right_roots), 'go-')

plt.axis([-4,4,-4,4])
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid(True)
plt.axis('equal')
plt.title('Complex numbers')
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')

plt.show()

args = np.angle(n_roots)
print(args)
pi_over_three = np.array( [pi/3]*8 )
print("First: dots, that are connected with blue stripe")
print("Second: dots, taht are shown with stars")
print("Third:",(abs(args) > pi_over_three))
print("P.S. dots are counted counterclockwise from one, that in center of first quadrant")
```

```python
# ЛОГИЧЕСКИЕ ОПЕРАЦИИ
import numpy as np

x = np.array( [0,0,1,1], dtype=bool)
y = np.array( [0,1,0,1], dtype=bool)
t = np.array( [1 < 2, 5 > 3] )

print('x = ' + str(x))
print('y = ' + str(y))
print('\nx and y = ' + str(x & y))
print('x and not y = ' + str(x & ~y))
print('x or y = ' + str(x | y))
print('x xor y = ' + str(x^y))
print('not x = ' + str(~x))
```

```python
# ПОСТРОЕНИЕ МНОЖЕСТВА ТОЧЕК КОМПЛ. ПЛ. УДОВЛ. НЕРАВЕНСТВУ
import numpy as np
import numpy.random as rng
import matplotlib.pyplot as plt

x = 3 - 6*rng.rand(10_000) # 10_000 точек х на [-3; 3]
y = 3 - 6*rng.rand(10_000)
z = x + 1j*y
L = (np.abs(z + 1j) >= 1.5)
plt.plot(x[L], y[L], '.')

plt.axis('square')
plt.grid(True)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.title('$| z+i | \geq 1.5 $')
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
```

```python
# УПРАЖНЕНИЕ 5
import numpy as np
import numpy.random as rng
import matplotlib.pyplot as plt

x = 4 - 8*rng.rand(100_000)
y = 4 - 8*rng.rand(100_000)
z = x + 1j*y
L1 = (np.abs(z - 1 - 1j) <= 2)
L2 = (1 <= np.abs(z - 1 - 1j))
L = L1 & L2

plt.plot(x[L], y[L], '.')

plt.axis('square')
plt.grid(True)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.title('$1 \geq | z-1-i | \geq 2 $')
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
```

```python
# УПРАЖНЕНИЕ С1
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt

n = -256 + 0j
r = np.abs(n)
phi = np.angle(n)
k = np.arange(1,9,1)
n_roots = r**(1/8) * (np.cos((phi + 2*np.pi*k) / 8) +
                   1j*np.sin((phi + 2*np.pi*k) / 8) )

n_args = np.angle(n_roots)
cond = (np.abs(n_args) > (np.pi/2))
n_roots_left = n_roots[cond]

cond = (np.isclose(np.imag(n_roots), 0))
n_roots_real = n_roots[cond]

cond = (np.abs(n_args) > (2*np.pi/3))
n_roots_2pi_over3 = n_roots[cond]

y1, = plt.plot(np.real(n_roots), np.imag(n_roots), 
               '*g', label='Остальные корни')
y2, = plt.plot(np.real(n_roots_left), np.imag(n_roots_left), 
               'or', label='Корни левой полуплоскости')
y3, = plt.plot(np.real(n_roots_real), np.imag(n_roots_real), 
               'ob', label='Корни на действ. оси')
y4, = plt.plot(np.real(n_roots_2pi_over3), np.imag(n_roots_2pi_over3), 
               'oc', label='Корни левой полуплоскости\nКорни с arg > $2*pi/3$')
plt.grid(True)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.legend(handles = [y1, y2, y4, y3])
plt.title('$root[8](-256)$')
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
```
