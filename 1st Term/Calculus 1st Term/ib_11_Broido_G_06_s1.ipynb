```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sympy.abc import x
func = x**2 - 4
sp.plot(func, (x, -5, 5))
sp.plot(func, (x, -5, 5), adaptive=False, nb_of_points=5)
# откючаем адаптивное кол-во точек, указываем их кол-во сами
```

```python
# ПРИМЕР 1
sp.plot( sp.cos(x), (x, 0, 2*sp.pi) )
y = sp.cos(x)
sp.plot( y, (x, 0, 2*sp.pi) )
sp.plot( y, (x, 0, 2*sp.pi), adaptive=False, nb_of_points=10,
        line_color='r')
```

```python
y = sp.sin(x) / (x**2 + 1)
sp.plot(sp.sin(x) / (x**2 + 1), (x,-20,20))
sp.plot(y, (x, -20, 20))
```

```python
# ПРИМЕР 2
from matplotlib import style

style.use('ggplot') # отображение сетки координат по умолчанию
y = sp.Abs(x)/x
sp.plot((y, (x,-1,0)), (y, (x,0,1)))
sp.plot(y, (x,-1,1))
```

```python
# УПРАЖНЕНИЕ 1
y1 = (2*x - 1) / (x - 1)
y2 = (2*x - 1) / abs(x - 1)

sp.plot((y1, (x, -5, 1)), (y1, (x, 1, 2)), (y1, (x, 2, 5)))
print("lim(x->+inf) (y1) = 2")
print("lim(x->-inf) (y1) = 2")
print("lim(x->1+0) (y1) = +inf")
print("lim(x->1-0) (y1) = -inf")
print("lim(x->1) (y1) = inf")
print("lim(x->2+-0) (y1) = 3")
sp.plot((y2, (x, -5, 1)), (y2, (x, 1, 2)), (y2, (x, 2, 5)))
print("lim(x->+inf) (y2) = 2")
print("lim(x->-inf) (y2) = -2")
print("lim(x->1+0) (y2) = +inf")
print("lim(x->1-0) (y2) = +inf")
print("lim(x->1) (y2) = +inf")
print("lim(x->2+-0) (y2) = 3")
```

```python
# УПРАЖНЕНИЕ С1
y1 = sp.Abs( (2*x - 1) / (x - 1) )
y2 = sp.Abs(3*x - 6) / (x - 2)
sp.plot(y1, (x, 1.5, 2.5))
sp.plot((y1, (x, -5, 1)), (y1, (x, 1, 7)))
print("lim(x->+inf) (y1) = 2")
print("lim(x->-inf) (y1) = 2")
print("lim(x->1+-0) (y1) = +inf")
print("lim(x->2+-0) (y1) = 3")
sp.plot((y2, (x, -8,2)), (y2, (x, 2, 12)))
print("lim(x->+inf) (y2) = 3")
print("lim(x->-inf) (y2) = -3")
print("lim(x->1+-0) (y2) = -3")
print("lim(x->2-0) (y2) = -3")
print("lim(x->2+0) (y2) = 3")
```
