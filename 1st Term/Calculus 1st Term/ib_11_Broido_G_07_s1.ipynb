```python
# brentq(function, a, b)
from scipy import optimize
from math import sin, cos, pi

def f(x):
    return (cos(x) - sin(x))

root = optimize.brentq(f, 0, pi/2)
print('x =', root)
print('f(x) =', f(root))
```

```python
from math import tan
def f(x):
    return (tan(x))
root = optimize.brentq(f, 1, 2)
print('x =', root)
print('f(x) =', f(root))
```

```python
# ПРИМЕР 6
from sympy.solvers import nsolve
from sympy import Symbol, sin

x = Symbol('x')
eq = sin(x) - x - 1
a = nsolve(eq, 0)

print('x =', a)
print('f(x) =', eq.subs(x,a)) # Подставили вместо х значение, в котором функция равна нулю (решение)
```

```python
# УПРАЖНЕНИЕ 3
# a)
from sympy.solvers import solve, nsolve
from sympy import Symbol, sin
from math import e

x = Symbol('x')
y = e**(3*x) - (3*x)**e
ans_n = nsolve(y, 0.95)
sp.plot(y, (x, 0.875,0.92))
print('x =', ans_n)
print('f(x) =', y.subs(x, ans_n))

# б)
y = sin(x) - x/3

sp.plot(y, (x,0, sp.pi))
ans1 = nsolve(y, 0)
ans2 = nsolve(y, 2.2)
print('x1, x2 =', ans1, ans2)
print('f(x1), f(x2) =', y.subs(x, ans1), y.subs(x, ans2))
```

```python
# РЕШЕНИЕ СИСТЕМЫ УРАВНЕНИЙ
# ПРИМЕР 8
from sympy.solvers import nsolve
from sympy import Symbol
x = Symbol('x')
y = Symbol('y')
y1 = x**2 - 2
y2 = 2*x + 5
sp.plot( (y1, (x, -3, 4)), (y2, (x, -3, 4)) )

f1 = x**2 - y - 2
f2 = 2*x - y + 5
a = nsolve( (f1, f2), (x,y), (0,0))
print('x = ', a)

a = nsolve( (f1, f2), (x,y), (3, 12))
print('x = ', a)
```

```python
# УПРАЖНЕНИЕ С3
# а)
from sympy.solvers import nsolve
from sympy import Symbol
x = Symbol('x')
f = x + 1/x - 2
ans = nsolve(f, 2)
print('x0 =', ans)
print('f(x0) =', f.subs(x, ans))
ans = nsolve(f, 1)
print('x0 =', ans)
print('f(x0) =', f.subs(x, ans))
```
