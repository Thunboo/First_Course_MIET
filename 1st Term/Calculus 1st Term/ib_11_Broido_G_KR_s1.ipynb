```python
import numpy as np
import numpy.random as rng
import matplotlib.pyplot as plt
```

```python
Z1_2 = np.array([-1 + np.sqrt(3)*1j, 0+2j])
Z3_6 = np.array([Z1_2[0] + Z1_2[1],
                 Z1_2[0] * Z1_2[1],
                 (Z1_2[0])**(1/3),
                 Z1_2[0]/Z1_2[1]]) 

pi = np.pi
r = np.abs(Z1_2[0])
phi = np.angle(Z1_2[0])
k = np.arange(1,4)
z_roots = r**(1/3) * ( np.cos( (phi + 2.0*pi*k) / 3) +
                  1j * np.sin( (phi + 2.0*pi*k) / 3))

orig = np.zeros(4)
plt.quiver(orig,orig, np.real(Z1_2[0]), np.imag(Z1_2[0]), angles='xy', 
               scale_units='xy', scale=1, color="blue")
plt.quiver(orig,orig, np.real(Z1_2[1]), np.imag(Z1_2[1]), angles='xy', 
               scale_units='xy', scale=1, color="red")
plt.scatter(np.real(Z3_6[0]),np.imag(Z3_6[0]), marker="+", color="green", s=100)
plt.scatter(np.real(Z3_6[1]),np.imag(Z3_6[1]), marker="*", color="blue", s=50)
for z in z_roots:    
    plt.scatter(np.real(z),np.imag(z), marker="^", color="red", s=50)
plt.scatter(np.real(Z3_6[3]),np.imag(Z3_6[3]), marker="*", color="red", s=50)
plt.axis([-4,4,-4,4])      
plt.axhline(y=0, color='k') 
plt.axvline(x=0, color='k')
plt.grid(True)
plt.title('Complex numbers')
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
```

```python
import numpy as np
import numpy.random as rng
import matplotlib.pyplot as plt

x = 4 - 8*rng.rand(100000)
y = 4 - 8*rng.rand(100000)
z = x + 1j*y
Z1, Z2 = -1 + np.sqrt(3)*1j, 0+2j
L1 = (np.abs(z - Z1) <= 2)
L2 = (np.abs(z + np.abs(Z2)) <= 2)
L = L1 | L2

plt.plot(x[L], y[L], '.')

plt.axis('square')
plt.grid(True)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.title('$ Второе задание $')
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
```

```python
import sympy as sp
from sympy import Symbol
n = Symbol('n')
xn = (n**2 - n)/(2*n**2 - 2*n - 1)
l = sp.Limit( xn, n, sp.oo) # , '-')
a = sp.limit( xn, n, sp.oo)
sp.pprint(l)
print('===')
sp.pprint(a)
###############################
#  ∀ ε>0 ∃ A(ε)>0 : |n| > A(ε) => |x_n - 1/2| < ε
print("∀ ε>0 ∃ A(ε)>0 : n > A(ε) => |x_n - 1/2| < ε")
###############################
e1 = sp.solve_univariate_inequality(sp.Abs(
                                    (n**2 - n)/(2*n**2 - 2*n - 1) - a) 
                                    < 0.01, n, relational=False)
e2 = sp.solve_univariate_inequality(n > 0, n, relational=False)
In = sp.Intersection(e1,e2)
sp.pprint(In)
print("=> n(0.01 = 6)")
##############################
n = np.arange(3,10,1)
xn = (n**2 - n)/(2*n**2 - 2*n - 1)

@np.vectorize
def const_a(x):
    return a

plt.plot(n, xn)
x = np.linspace(2,11,100)
plt.plot(x, const_a(x))
plt.text(8+0.2, 0.5-0.02, '$A = 1/2$')
plt.scatter(6,0, color='red', s=60)
plt.text(6+0.2, 0-0.02, 'n_0(0.01)')
plt.grid()
plt.axhline(y = 0, color = 'k')
plt.axvline(x = 0, color = 'k')
plt.show()
```
