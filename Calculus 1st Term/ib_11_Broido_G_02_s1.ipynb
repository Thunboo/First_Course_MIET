```python
# Построение графика в декартовых координатах
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-2, 2, 1000) # Задаём отрезок значений х [-2, 2] разбитый на 1000 равных частей =>
                             # => более ровный график, а не кусочно линейная функция
y = np.exp(x) # Задаём зависимость y(x)
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(-2,2,1000)
y = np.exp(x)
plt.plot(x,y)
plt.xlabel('x') # Подписываем оси
plt.ylabel('y')
plt.title('Function $y = e^{x}$') # Название графика задаётся только перед командой plt.show() 
                                  # $ для отображения заданной зависимости в математической форме
plt.show()
```

```python
figure_1 = plt.figure()             # Добавляем графическое figure 1
ax1 = figure_1.add_subplot(1, 1, 1) # Добавляем график 1 на figure 1
plt.title('Function $y = e^{x}$')
ax1.plot(x,y)                       # Строим функию на графике
figure_1.show()

figure_2 = plt.figure()         
ax2 = figure_2.add_subplot(1, 1, 1) # (n,k,cur) => n рядов, k колонок для разбиения окна на подобласти
                                    # cur - порядковый номер подобласти в котором plot() будет строить график
plt.title('Function $y = ln(x)$')
ax2.plot(y,x)                       # Третий аргумент - цвет графика ( y, m, c, g, b, w, r, k)
figure_2.show()
```

```python
# УПРАЖНЕНИЕ 2
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-2*np.pi, 2*np.pi, 25)
y1, = plt.plot(x, np.sin(x), '^r--', label='y=sin(x)')          # Обазятельно указывать "," для отображения легенды
y2, = plt.plot(x, np.sin(x) - 2, 'xy-.', label='y=sin(x) - 2')  # Передаём переменные и названия функциям plot
y3, = plt.plot(x, np.sin(x) + 1, '.g-' , label='y=sin(x) + 1')  # Сохраняем ссылки на графический объект, 
                                                                   # возвращаемый  функцией plot
plt.legend(handles = [y1, y2, y3])  # Передаём ссылки функцие plt.legend()
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')

plt.xlabel('X value')
plt.ylabel('Y value')
plt.title('Sine functions')

plt.show()
```

```python
# УПРАЖНЕНИЕ 4
x = np.linspace(-5,5, 100)
def f(x):
    return abs( abs(x) - 2)
plt.subplot(2,3,1)
plt.plot(x, f(x))
plt.grid(True)

plt.subplot(2,3,2)
plt.plot(x, f(x-2))
plt.grid(True)

plt.subplot(2,3,3)
plt.plot(x, f(x+2))
plt.grid(True)

plt.subplot(2,3,4)
plt.plot(x, f(2*x))
plt.grid(True)

plt.subplot(2,3,5)
plt.plot(x, f(0.5*x))
plt.grid(True)

plt.subplot(2,3,6)
plt.plot(x, f(-x))
plt.grid(True)

```

```python
# УПРАЖНЕНИЕ 7
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, np.pi, 100)
y1, = plt.plot(x, np.cos(x), 'g:', label = '$y=cos(x)$')
y2, = plt.plot(x, np.arccos(x), 'r-.', label = '$y=arccos(x)$')
y_sym=plt.plot(x, x, 'b', '$y=x$')

plt.legend(handles = [y1, y2])
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.axis('equal')
plt.grid(True)

plt.xlabel(' X Value')
plt.ylabel(' Y Value ')
plt.title(' График cos и arccos, ось их симметрии ')

plt.show()
```

```python
# УПРАЖНЕНИЕ С3
import matplotlib.pyplot as plt
import numpy as np

plt.subplot(2,1,1)
x = np.linspace(-2, 2, 100)
y1, = plt.plot(x, x,                'r-', label = '$y = x$')
y2, = plt.plot(x, 1/x,              'g:', label = '$y = 1/x$')
y3, = plt.plot(x, 1 / (np.sqrt(x)), 'b--',label = '$y = 1/sqrt(x)$')

plt.legend(handles = [y1, y2, y3])
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
#plt.axis('equal')
plt.grid(True)
plt.xlabel(' X Value')
plt.ylabel(' Y Value ')
##################################################
plt.subplot(2,1,2)
x = np.linspace(-2, 2, 30)
y1, = plt.semilogy(x, x,                'r-', label = '$y = log(x)$')
y2, = plt.semilogy(x, 1/x,              'g:', label = '$y = log(1/x)$')
y3, = plt.semilogy(x, 1 / (np.sqrt(x)), 'b--',label = '$y = log(1/sqrt(x))$')

plt.legend(handles = [y1, y2, y3])
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
#plt.axis('equal')
plt.grid(True)

plt.xlabel(' X Value')
plt.ylabel(' Y Value ')
#plt.title(' ... ')

plt.show()
```
