```python
# ПРИМЕР 5
import matplotlib.pyplot as plt
import numpy as np
z = 1 + 1j
pi = np.pi
r = np.abs(z)
phi = np.angle(z)
k = np.arange(1,7)
z_roots = r**(1/6) * ( np.cos( (phi + 2.0*pi*k) / 6) +
                  1j * np.sin( (phi + 2.0*pi*k) / 6))
plt.title('Sixth roots of ${1+j}$')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid(True)
plt.axis('equal')
plt.plot(np.real(z_roots), np.imag(z_roots), '*g')

t = np.linspace(0, 2*pi, 100)
plt.plot(r**(1/6) * np.cos(t), r**(1/6)*np.sin(t), ':r')

q1 = plt.quiver(0,0, np.real(z), np.imag(z), angles = 'xy', 
           scale_units = 'xy', scale = 1.0, color = 'Black',
           label = '$1 + j$ quiver')
plt.legend( handles = [q1])

#print(z_roots)
#for i in z_roots:
#    print(i**6)
```

```python
# КОНТРОЛЬНЫЕ ВОПРОСЫ
#
# 1) Комплексное число можно ввести при помощи указания на мнимую часть
#   с помощью символа 'j' : (1 + 1j). Также, можно использовать метод
#   complex( *real part* , *imaginary part* )

# 2) Вещественная часть комплексного числа - число, занимающее ось абцисс
#   координатной плоскости и являющееся частью множества вещественных
#   чисел. Мнимой частью числа является число формата b*i (или j), где
#   i (или j) - мнимая единица - элемент, расширяющий множество 
#   вещественных числе до множества комплексных. Ось комплексных чисел
#   расположена перпендикулярно оси вещественных чисел

# 3) Арифметические действия над комплексными числами осуществляются
#   с помошью привычных символов : +,-,*,/,** и так далее

# 4) Модуль комплексного числа - длина вектора, представляющего собой
#   комплексное число в декартовой плоскости.
#    Аргумент комплексного числа - угол, на который отклонён вектор,
#   представляющий комплексное число, от положительного направления оси
#   абцисс

# 5) Модуль произведения (частного) двух комплексных чисел = произведению (частному) модулей этих чисел.
#    Аргумент произведения (частного) двух комплексных числе = сумме (разности) аргументов этих чисел.
 
# 6) У комплексного числа существует n корней n-ой степени. Они
#   располагаются равномерно по окружности с углом равным 2pi/n меж собой
```

```python
# УПРАЖНЕНИЕ С1
import numpy as np

# 1)
a = ((1 + 2j)**2) * (1 - 1j)
print(a)

# 2)
b = (1 - 1j)**4 - (1 + 1j)**4
print(b)

# 3)
x = np.array([1j]*21)
for i in range(20,41):
    x[i-20] = x[i-20]**i
print(x)

# 4)
c = ((2 + 5j) / (1 - 3j)) + ( 1 / (1 - 1j))
print(c)
c_abs = format(np.abs(c), '0.3f')
c_arg = format(np.angle(c), '0.3f')
print(f'~{c_abs} * e^(i*{c_arg})')
print(f'~{c_abs} * ( cos({c_arg}) + i*sin({c_arg}) )')
```

```python
# УПРАЖНЕНИЕ С3
import numpy as np
def find_parts_abs_arg_conj(cn):
    Re = np.real(cn)
    Im = np.imag(cn)
    Abs = np.abs(cn)
    Arg = np.angle(cn)
    Conj = np.conj(cn)
    return f'Real part: {Re}\nImaginary part: {Im}\nAbs: {Abs}\nArgument: {Arg}\nConjugate: {Conj}\n'

# 1)
a = (5 + 4j) * (3 - 2*(1j)**3)
print(find_parts_abs_arg_conj(a))
# 2)
b = (1- 1j)**13
print(find_parts_abs_arg_conj(b))
```
